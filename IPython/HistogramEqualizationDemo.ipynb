# Histogram Equalization Demo

The purpose of this code is implement HDR tonemapping using histogram equalization. Our steps are as follows:

1. tranform an image into xyY space
1. compute a histogram using atomic operations
1. compute the cumulative distribution function of the histogram using data reductions and prefix sum operations
1. compute the final luminance by indexing into the cumulative distribution function
1. transform the image back into RGB space

See [1-3] for details. 

[1] http://radsite.lbl.gov/radiance/papers/lbnl39882/tonemap.pdf

[2] http://www.cs.utah.edu/~reinhard/cdrom/tonemap.pdf

[3] http://developer.amd.com/gpu_assets/GPUHistogramGeneration_I3D07.pdf

## load HDR image

```python
import array
import OpenEXR
import Imath

pt         = Imath.PixelType(Imath.PixelType.FLOAT)
rgb_hdr    = OpenEXR.InputFile("Memorial.exr")
header     = rgb_hdr.header()
dw         = header["dataWindow"]
sz         = (dw.max.x - dw.min.x + 1, dw.max.y - dw.min.y + 1)

r_hdr, g_hdr, b_hdr = rgb_hdr.channels("RGB", pt)

r = numpy.fromstring(r_hdr, dtype=numpy.float32).reshape((sz[1], sz[0]))
g = numpy.fromstring(g_hdr, dtype=numpy.float32).reshape((sz[1], sz[0]))
b = numpy.fromstring(b_hdr, dtype=numpy.float32).reshape((sz[1], sz[0]))

brightest = max(numpy.max(r), numpy.max(g), numpy.max(b))
r, g, b   = r / brightest, g / brightest, b / brightest
rgb       = numpy.concatenate((r[:,:,newaxis],g[:,:,newaxis],b[:,:,newaxis]), axis=2).copy()



figsize(22, 4)

matplotlib.pyplot.subplot(161);
matplotlib.pyplot.imshow(numpy.clip(200 * rgb, 0, 1));
matplotlib.pyplot.title("numpy.clip(200 * rgb, 0, 1)");

matplotlib.pyplot.subplot(162);
matplotlib.pyplot.imshow(numpy.clip(400 * rgb, 0, 1));
matplotlib.pyplot.title("numpy.clip(400 * rgb, 0, 1)");

matplotlib.pyplot.subplot(163);
matplotlib.pyplot.imshow(numpy.clip(800 * rgb, 0, 1));
matplotlib.pyplot.title("numpy.clip(800 * rgb, 0, 1)");

matplotlib.pyplot.subplot(164);
matplotlib.pyplot.imshow(numpy.clip(1600 * rgb, 0, 1));
matplotlib.pyplot.title("numpy.clip(1600 * rgb, 0, 1)");

matplotlib.pyplot.subplot(165);
matplotlib.pyplot.imshow(numpy.clip(3200 * rgb, 0, 1));
matplotlib.pyplot.title("numpy.clip(3200 * rgb, 0, 1)");

matplotlib.pyplot.subplot(166);
matplotlib.pyplot.imshow(numpy.clip(6400 * rgb, 0, 1));
matplotlib.pyplot.title("numpy.clip(6400 * rgb, 0, 1)");
```

## perform log-luminance histogram equalization on the CPU

```python
X = r * 0.4124 + g * 0.3576 + b * 0.1805
Y = r * 0.2126 + g * 0.7152 + b * 0.0722
Z = r * 0.0193 + g * 0.1192 + b * 0.9505

L = X + Y + Z
x = X / L
y = Y / L

delta          = 0.0001

log_Y          = numpy.log10(delta + Y)
min_log_Y      = numpy.min(log_Y)
max_log_Y      = numpy.max(log_Y)
log_Y_range    = max_log_Y - min_log_Y
    
num_bins       = 1024
hist, edges    = numpy.histogram(log_Y, num_bins)

cdf            = numpy.cumsum(hist)
cdf_total      = cdf[-1]
cdf_norm       = cdf.astype(numpy.float32) / cdf_total

bin_indices    = ( (num_bins * (log_Y - min_log_Y)) / log_Y_range ).astype(numpy.int32) - 1

log_Y_new_norm = cdf_norm[bin_indices]
Y_new          = log_Y_new_norm

X_new = x * (Y_new / y)
Z_new = (1 - x - y) * (Y_new / y);

r_new_cpu = X_new *  3.2406 + Y_new * -1.5372 + Z_new * -0.4986;
g_new_cpu = X_new * -0.9689 + Y_new *  1.8758 + Z_new *  0.0415;
b_new_cpu = X_new *  0.0557 + Y_new * -0.2040 + Z_new *  1.0570;

rgb_new_cpu = numpy.concatenate((r_new_cpu[:,:,newaxis],g_new_cpu[:,:,newaxis],b_new_cpu[:,:,newaxis]), axis=2).copy()



figsize(14, 9)

matplotlib.pyplot.subplot(231);
matplotlib.pyplot.imshow(numpy.clip(1000 * rgb, 0, 1));
matplotlib.pyplot.title("numpy.clip(1000 * rgb, 0, 1)");

matplotlib.pyplot.subplot(232);
matplotlib.pyplot.imshow(Y, cmap="gray");
matplotlib.pyplot.title("Y");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(233);
matplotlib.pyplot.imshow(log_Y, cmap="gray");
matplotlib.pyplot.title("log_Y");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(234);
matplotlib.pyplot.imshow(bin_indices, cmap="gray");
matplotlib.pyplot.title("bin_indices");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(235);
matplotlib.pyplot.imshow(Y_new, cmap="gray");
matplotlib.pyplot.title("Y_new");
matplotlib.pyplot.colorbar();

matplotlib.pyplot.subplot(236);
matplotlib.pyplot.imshow(numpy.clip(rgb_new_cpu, 0, 1));
matplotlib.pyplot.title("numpy.clip(rgb_new_cpu, 0, 1)");
```

## allocate GPU memory

```python
import pycuda.autoinit
import pycuda.driver
import pycuda.compiler
import reduce
import prefixsum

num_pixels_y          = numpy.int32(r.shape[0])
num_pixels_x          = numpy.int32(r.shape[1])
num_bins              = 1024
size_of_element_bytes = 4

r_gpu = numpy.zeros_like(r, dtype=numpy.float32)
g_gpu = numpy.zeros_like(g, dtype=numpy.float32)
b_gpu = numpy.zeros_like(b, dtype=numpy.float32)

r_new_gpu = numpy.zeros_like(r, dtype=numpy.float32)
g_new_gpu = numpy.zeros_like(g, dtype=numpy.float32)
b_new_gpu = numpy.zeros_like(b, dtype=numpy.float32)

r_device = pycuda.driver.mem_alloc(r.nbytes)
g_device = pycuda.driver.mem_alloc(r.nbytes)
b_device = pycuda.driver.mem_alloc(r.nbytes)

x_device     = pycuda.driver.mem_alloc(r.nbytes)
y_device     = pycuda.driver.mem_alloc(r.nbytes)
Y_device     = pycuda.driver.mem_alloc(r.nbytes)
log_Y_device = pycuda.driver.mem_alloc(r.nbytes)

hist_device     = pycuda.driver.mem_alloc(num_bins * size_of_element_bytes)
cdf_device      = pycuda.driver.mem_alloc(num_bins * size_of_element_bytes)
cdf_norm_device = pycuda.driver.mem_alloc(num_bins * size_of_element_bytes)

r_new_device = pycuda.driver.mem_alloc(r_new_gpu.nbytes)
g_new_device = pycuda.driver.mem_alloc(r_new_gpu.nbytes)
b_new_device = pycuda.driver.mem_alloc(r_new_gpu.nbytes)

reduce_manager     = reduce.ReduceManager(1024 * 1024)
prefix_sum_manager = prefixsum.PrefixSumManager(2048)
```

## define GPU kernels

```python
source_module = pycuda.compiler.SourceModule \
(
"""
__global__ void rgb_to_xyY(
    float* d_r,
    float* d_g,
    float* d_b,
    float* d_x,
    float* d_y,
    float* d_Y,
    float* d_log_Y,
    float  delta,
    int    num_pixels_y,
    int    num_pixels_x )
{
    int  ny             = num_pixels_y;
    int  nx             = num_pixels_x;
    int2 image_index_2d = make_int2( ( blockIdx.x * blockDim.x ) + threadIdx.x, ( blockIdx.y * blockDim.y ) + threadIdx.y );
    int  image_index_1d = ( nx * image_index_2d.y ) + image_index_2d.x;

    if ( image_index_2d.x < nx && image_index_2d.y < ny )
    {
        float r = d_r[ image_index_1d ];
        float g = d_g[ image_index_1d ];
        float b = d_b[ image_index_1d ];

        float X = ( r * 0.4124 ) + ( g * 0.3576 ) + ( b * 0.1805 );
        float Y = ( r * 0.2126 ) + ( g * 0.7152 ) + ( b * 0.0722 );
        float Z = ( r * 0.0193 ) + ( g * 0.1192 ) + ( b * 0.9505 );

        float L = X + Y + Z;
        float x = X / L;
        float y = Y / L;

        float log_Y = log10( delta + Y );

        d_x[ image_index_1d ]     = x;
        d_y[ image_index_1d ]     = y;
        d_Y[ image_index_1d ]     = Y;
        d_log_Y[ image_index_1d ] = log_Y;
    }
}

__global__ void compute_histogram(
    unsigned int* d_histogram,
    float*        d_log_Y,
    float         min_log_Y,
    float         max_log_Y,
    float         log_Y_range,
    int           num_bins,
    int           num_pixels_y,
    int           num_pixels_x )
{
    int  ny             = num_pixels_y;
    int  nx             = num_pixels_x;
    int2 image_index_2d = make_int2( ( blockIdx.x * blockDim.x ) + threadIdx.x, ( blockIdx.y * blockDim.y ) + threadIdx.y );
    int  image_index_1d = ( nx * image_index_2d.y ) + image_index_2d.x;

    if ( image_index_2d.x < nx && image_index_2d.y < ny )
    {
        float log_Y     = d_log_Y[ image_index_1d ];
        int   bin_index = min( num_bins - 1, int( ( num_bins * ( log_Y - min_log_Y ) ) / log_Y_range ) );

        atomicAdd( d_histogram + bin_index, 1 );
    }
}

__global__ void normalize_cdf(
    unsigned int* d_input_cdf,
    float*        d_output_cdf,
    float         normalization_constant,
    int           n
)
{
    int global_index_1d = ( blockIdx.x * blockDim.x ) + threadIdx.x;

    if ( global_index_1d < n )
    {
        unsigned int input_value  = d_input_cdf[ global_index_1d ];
        float        output_value = input_value * normalization_constant;

        d_output_cdf[ global_index_1d ] = output_value;
    }
}

__global__ void tonemap(
    float* d_x,
    float* d_y,
    float* d_log_Y,
    float* d_cdf_norm,
    float* d_r_new,
    float* d_g_new,
    float* d_b_new,
    float  min_log_Y,
    float  max_log_Y,
    float  log_Y_range,
    int    num_bins,
    int    num_pixels_y,
    int    num_pixels_x )
{
    int  ny             = num_pixels_y;
    int  nx             = num_pixels_x;
    int2 image_index_2d = make_int2( ( blockIdx.x * blockDim.x ) + threadIdx.x, ( blockIdx.y * blockDim.y ) + threadIdx.y );
    int  image_index_1d = ( nx * image_index_2d.y ) + image_index_2d.x;

    if ( image_index_2d.x < nx && image_index_2d.y < ny )
    {
        float x         = d_x[ image_index_1d ];
        float y         = d_y[ image_index_1d ];
        float log_Y     = d_log_Y[ image_index_1d ];
        int   bin_index = min( num_bins - 1, int( (num_bins * ( log_Y - min_log_Y ) ) / log_Y_range ) );
        float Y_new     = d_cdf_norm[ bin_index ];

        float X_new = x * ( Y_new / y );
        float Z_new = ( 1 - x - y ) * ( Y_new / y );
        
        float r_new = ( X_new *  3.2406 ) + ( Y_new * -1.5372 ) + ( Z_new * -0.4986 );
        float g_new = ( X_new * -0.9689 ) + ( Y_new *  1.8758 ) + ( Z_new *  0.0415 );
        float b_new = ( X_new *  0.0557 ) + ( Y_new * -0.2040 ) + ( Z_new *  1.0570 );

        d_r_new[ image_index_1d ] = r_new;
        d_g_new[ image_index_1d ] = g_new;
        d_b_new[ image_index_1d ] = b_new;
    }
}
"""
)

rgb_to_xyY_function              = source_module.get_function("rgb_to_xyY")
compute_histogram_function       = source_module.get_function("compute_histogram")
normalize_cdf_function           = source_module.get_function("normalize_cdf")
tonemap_function                 = source_module.get_function("tonemap")

delta                            = numpy.float32(0.0001)
num_pixels_y                     = numpy.int32(r.shape[0])
num_pixels_x                     = numpy.int32(r.shape[1])

rgb_to_xyY_function_block        = (32,16,1)
num_blocks_y                     = int(ceil(float(num_pixels_y) / float(rgb_to_xyY_function_block[1])))
num_blocks_x                     = int(ceil(float(num_pixels_x) / float(rgb_to_xyY_function_block[0])))
rgb_to_xyY_function_grid         = (num_blocks_x, num_blocks_y)

compute_histogram_function_block = (32,16,1)
num_blocks_y                     = int(ceil(float(num_pixels_y) / float(compute_histogram_function_block[1])))
num_blocks_x                     = int(ceil(float(num_pixels_x) / float(compute_histogram_function_block[0])))
compute_histogram_function_grid  = (num_blocks_x, num_blocks_y)

normalize_cdf_function_block     = (512,1,1)
num_blocks                       = int(ceil(float(num_bins) / float(normalize_cdf_function_block[0])))
normalize_cdf_function_grid      = (num_blocks, 1)

tonemap_function_block           = (32,16,1)
num_blocks_y                     = int(ceil(float(num_pixels_y) / float(tonemap_function_block[1])))
num_blocks_x                     = int(ceil(float(num_pixels_x) / float(tonemap_function_block[0])))
tonemap_function_grid            = (num_blocks_x, num_blocks_y)
```

```python
pycuda.driver.memcpy_htod(r_device, r)
pycuda.driver.memcpy_htod(g_device, g)
pycuda.driver.memcpy_htod(b_device, b)

pycuda.driver.memset_d32(hist_device, int(0), num_bins)

rgb_to_xyY_function(
    r_device,
    g_device,
    b_device,
    x_device,
    y_device,
    Y_device,
    log_Y_device,
    delta,
    num_pixels_y,
    num_pixels_x,
    block=rgb_to_xyY_function_block,
    grid=rgb_to_xyY_function_grid)

min_log_Y   = reduce_manager.reduce_min_device(log_Y_device, num_pixels_y * num_pixels_x)
max_log_Y   = reduce_manager.reduce_max_device(log_Y_device, num_pixels_y * num_pixels_x)
log_Y_range = max_log_Y - min_log_Y

compute_histogram_function(
    hist_device,
    log_Y_device,
    numpy.float32(min_log_Y),
    numpy.float32(max_log_Y),
    numpy.float32(log_Y_range),
    numpy.int32(num_bins),
    num_pixels_y,
    num_pixels_x,
    block=compute_histogram_function_block,
    grid=compute_histogram_function_grid)

prefix_sum_manager.prefix_sum_device(hist_device, cdf_device, num_bins)

tmp = numpy.zeros(1, dtype=numpy.uint32)
pycuda.driver.memcpy_dtoh(tmp, int(cdf_device) + (num_bins - 1) * size_of_element_bytes)
cdf_device_total = tmp[0]

normalize_cdf_function(
    cdf_device,
    cdf_norm_device,
    numpy.float32(1.0 / cdf_device_total),
    numpy.int32(num_bins),
    block=normalize_cdf_function_block,
    grid=normalize_cdf_function_grid)

tonemap_function(
    x_device,
    y_device,
    log_Y_device,
    cdf_norm_device,
    r_new_device,
    g_new_device,
    b_new_device,
    numpy.float32(min_log_Y),
    numpy.float32(max_log_Y),
    numpy.float32(log_Y_range),
    numpy.int32(num_bins),
    num_pixels_y,
    num_pixels_x,
    block=tonemap_function_block,
    grid=tonemap_function_grid)

pycuda.driver.memcpy_dtoh(r_new_gpu, r_new_device)
pycuda.driver.memcpy_dtoh(g_new_gpu, g_new_device)
pycuda.driver.memcpy_dtoh(b_new_gpu, b_new_device)

rgb_new_gpu = numpy.concatenate((r_new_gpu[:,:,newaxis],g_new_gpu[:,:,newaxis],b_new_gpu[:,:,newaxis]), axis=2).copy()



figsize(9, 4)

matplotlib.pyplot.subplot(121);
matplotlib.pyplot.imshow(numpy.clip(1000 * rgb, 0, 1));
matplotlib.pyplot.title("numpy.clip(1000 * rgb, 0, 1)");

matplotlib.pyplot.subplot(122);
matplotlib.pyplot.imshow(numpy.clip(rgb_new_gpu, 0, 1));
matplotlib.pyplot.title("numpy.clip(rgb_new_gpu, 0, 1)");
```

## verify correctness

```python
diff = \
    numpy.abs(r_new_gpu - r_new_cpu) + \
    numpy.abs(g_new_gpu - g_new_cpu) + \
    numpy.abs(b_new_gpu - b_new_cpu)

max_diff = numpy.ones_like(diff, dtype=numpy.float32) * 3



print \
    "Difference between GPU and CPU results as a percentage of the maximum possible difference (should be less than 1%%): %0.2f%%" % \
    (100 * (numpy.linalg.norm(diff) / numpy.linalg.norm(max_diff)))
    
print

figsize(15, 4)

matplotlib.pyplot.subplot(141);
matplotlib.pyplot.imshow(numpy.clip(rgb * 1000, 0, 1));
matplotlib.pyplot.title("numpy.clip(rgb * 1000, 0, 1)");

matplotlib.pyplot.subplot(142);
matplotlib.pyplot.imshow(numpy.clip(rgb_new_cpu, 0, 1));
matplotlib.pyplot.title("numpy.clip(rgb_new_cpu, 0, 1)");

matplotlib.pyplot.subplot(143);
matplotlib.pyplot.imshow(numpy.clip(rgb_new_gpu, 0, 1));
matplotlib.pyplot.title("numpy.clip(rgb_new_gpu, 0, 1)");

matplotlib.pyplot.subplot(144);
matplotlib.pyplot.imshow(diff, cmap="gray");
matplotlib.pyplot.title("diff");
matplotlib.pyplot.colorbar();
```
